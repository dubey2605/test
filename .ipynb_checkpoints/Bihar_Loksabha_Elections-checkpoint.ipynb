```python
result_df = final_df.groupby(['Party', 'Year'])['Voting %'].value_counts().reset_index()
result_df = result_df.drop(columns=['count'])
```

```python
result_df
```

```python
result_df = result_df[~((result_df['Party'] == 'Nationalist Congress Party') & (result_df['Year'] == 2014))]
```

```python
filter1= result_df[result_df['Party']=='Independent']
```

```python
filter1 = result_df['Party'] == 'Independent'
result_df = result_df[~filter1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set(style="whitegrid")

# Create a grouped bar plot
plt.figure(figsize=(14, 6))
ax = sns.barplot(x='Party', y='Voting %', hue='Year', data=result_df, palette='bright')

# Set labels and title
plt.xlabel('Political Party')
plt.ylabel('Voting Percentage')
plt.title('Bihar Lok Sabha Elections: Party-wise Voting % (2009-2019)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Annotate each bar with its height
for p in ax.patches:
    height = p.get_height()
    plt.annotate(f'{height:.0f}%', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Move the legend to the right
plt.legend(loc='upper right', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
# Define dictionary mapping regions to states
region_mapping = {
    'Western': ['Rajasthan', 'Maharashtra', 'Gujarat', 'Goa', 'Dadra_&_Nagar_Haveli, _Daman_&_Diu','Dadra & Nagar Haveli And Daman & Diu','Daman_&_Diu'],
    'Central': ['Madhya_Pradesh', 'Chhattisgarh'],
    'Northern': ['Uttar_Pradesh', 'Jammu_&_Kashmir', 'Himachal_Pradesh', 'Punjab', 'Uttarakhand', 'Haryana', 'Delhi', 'Chandigarh'],
    'Southern': ['Karnataka', 'Andhra_Pradesh', 'Telangana', 'Tamil_Nadu', 'Kerala', 'Lakshadweep', 'Andaman_&_Nicobar_Islands', 'Puducherry','Mysore','Madras'],
    'Eastern': ['Odisha', 'West_Bengal', 'Sikkim', 'Assam', 'Arunachal_Pradesh', 'Meghalaya', 'Manipur', 'Mizoram', 'Nagaland', 'Tripura', 'Bihar', 'Jharkhand'],
}


# Define a function to map state names to regions
def map_state_to_region(state):
    for region, states in region_mapping.items():
        if state in states:
            return region
    return None  # Return None if state is not found in any region

# Map 'State_Name' to 'Region' using the map_state_to_region function
new_df['Region'] = new_df['State_Name'].apply(lambda x: map_state_to_region(x))
```

```python
contested_winners_2019=new_df.query("(Position == 1) & (Year == 2019) & (Party == 'BJP')").groupby(['Region']).size().reset_index(name='Count')
```

```python
# Filter rows where BJP contested but didn't win (Position > 1)
bjp_contested = new_df.query("(Party == 'BJP') & (Position > 0)& (Year==2019)").groupby('Region').size().reset_index(name='Contested_Seats')

# Merge the contested seats count with the previous DataFrame
result_df = pd.merge(contested_winners_2019, bjp_contested, on='Region', how='left')
#result_df.fillna(0, inplace=True)  # Fill NaN values with 0 for regions where BJP didn't contest any seat

# Rename the columns for clarity
result_df.rename(columns={'Count': 'Seats_Won'}, inplace=True)
```

```python
women_mp=newest_df.query("Position == 1 & Sex == 'F' & Election_Type == 'Lok Sabha Election (GE)'& Year in(1967, 1971, 1977, 1980, 1984, 1989, 1991, 1996, 1998, 1999, 2004, 2009, 2014, 2019)").groupby(['Year']).size().reset_index(name='Count')
```

```python
import pandas as pd

# Example DataFrame
data = {
    'Year': [1962, 1967, 1971, 1977, 1980, 1984, 1989, 1991, 1996, 1998, 1999, 2004, 2009, 2014, 2019],
    'Count': [36, 32, 27, 19, 31, 44, 31, 40, 43, 45, 51, 47, 58, 65, 78]
}
women_mp = pd.DataFrame(data)

# Define a function to calculate the percentage based on the year
def calculate_percentage(row):
    total_seats = 543 if row['Year'] > 1962 else 494  # Conditionally set total seats
    return (row['Count'] / total_seats) * 100

# Apply the function to each row
women_mp['Percentage'] = women_mp.apply(calculate_percentage, axis=1)

# Display the updated DataFrame to verify the new column
print(women_mp)
```

```python
women_mp
```

```python
women_mp['Percentage']=(women_mp['Count']/543)*100
```

```python
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Convert 'Year' to string for better x-axis tick formatting
women_mp['Year'] = women_mp['Year'].astype(str)

# First plot with 'Count' on the left y-axis
color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Count', color=color)
ax1.plot(women_mp['Year'], women_mp['Count'], color=color, marker='o', linestyle='-')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

# Create a second axes that shares the same x-axis
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Percentage', color=color)
ax2.plot(women_mp['Year'], women_mp['Percentage'], color=color, marker='o', linestyle='-')
ax2.tick_params(axis='y', labelcolor=color)

# Adding title and customizing layout
plt.title('Number of Women MPs and Their Percentage in Lok Sabha Over the Years', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
fig.tight_layout()  # Adjust layout to prevent clipping of ylabel

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Styling improvements
ax1.set_xlabel('Year', fontsize=14, fontweight='bold')
ax1.set_ylabel('Count', fontsize=14, color='navy', fontweight='bold')
ax1.plot(women_mp['Year'], women_mp['Count'], color='navy', marker='o', linestyle='-', linewidth=2, label='Count')
ax1.tick_params(axis='y', labelcolor='navy')
ax1.grid(True, linestyle='--', alpha=0.5)

# Create a second axes that shares the same x-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Percentage', fontsize=14, color='crimson', fontweight='bold')
ax2.plot(women_mp['Year'], women_mp['Percentage'], color='crimson', marker='s', linestyle='-.', linewidth=2, label='Percentage')
ax2.tick_params(axis='y', labelcolor='crimson')

# Adding title and customizing layout
plt.title('Number of Women MPs and Their Percentage in Lok Sabha Over the Years', fontsize=16, fontweight='bold')
plt.xticks(women_mp['Year'], rotation=45)
fig.tight_layout()

# Adding a legend that includes both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```
